```python
import pandas as pd
from pandas import Timestamp
from pytvlwcharts import *
from binance.client import Client
```

```python

# Initialize Binance client
client = Client()
# Fetch BTCUSDT candlestick data
klines = client.get_klines(symbol='BTCUSDT', interval=Client.KLINE_INTERVAL_12HOUR, limit=100)

# Prepare candlestick data
df = pd.DataFrame(klines, columns=['time', 'open', 'high', 'low', 'close', 'volume', 'close_time', 'quote_asset_volume', 'number_of_trades', 'taker_buy_base_asset_volume', 'taker_buy_quote_asset_volume', 'ignore'])
df['time'] = df['time']/1000
df = df[['time', 'open', 'high', 'low', 'close']]
df = df.astype({'open': 'float', 'high': 'float', 'low': 'float', 'close': 'float'})

# Plot
DailyChart = Chart(
    data=df,
    width = 1295,
    height = 500,
    time_scale=TimeScaleOptions(
        seconds_visible=True,
        time_visible=True)
    ).mark_candlestick()
DailyChart
```
